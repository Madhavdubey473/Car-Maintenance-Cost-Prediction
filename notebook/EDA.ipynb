```python
# define numerical & categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
# Categorical Count Plots
for col in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Numerical Histograms & Boxplots
for col in numeric_features:
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(df[col], kde=True, ax=ax[0])
    ax[0].set_title(f'Distribution of {col}')
    sns.boxplot(x=df[col], ax=ax[1])
    ax[1].set_title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()
```

```python
# Categorical vs Cost
for col in categorical_features:
    plt.figure(figsize=(10, 5))
    sns.boxplot(data=df, x=col, y='cost')
    plt.title(f'Cost Distribution by {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Numerical vs Cost Scatterplots
for col in ['make_year', 'mileage', 'mileage_range']:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=df[col], y=df['cost'])
    plt.title(f'{col} vs Cost')
    plt.tight_layout()
    plt.show()
```

```python

# Correlation Heatmap
plt.figure(figsize=(12, 10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Pairplot for key features
sns.pairplot(df[['mileage', 'make_year', 'mileage_range', 'cost']])
plt.show()
```

```python

# ---------------------------
# 5. Binary Service Components vs Cost
# ---------------------------

# Calculate average cost when each component is changed
binary_features = df.columns[6:-1].difference(df[['mileage_range','cost','mileage']].columns)
component_costs = {}
for col in binary_features:
    avg_cost = df.groupby(col)['cost'].mean()
    component_costs[col] = avg_cost.get(1, 0)

```

```python
# Sort and plot
component_costs = pd.Series(component_costs).sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=component_costs.values, y=component_costs.index)
plt.title('Average Cost When Component is Replaced')
plt.xlabel('Average Cost')
plt.tight_layout()
plt.show()
```

```python
# ========== AGGREGATED ANALYSIS ==========
avg_cost_by_brand = df.groupby('brand')['cost'].mean().sort_values(ascending=False)
avg_cost_by_region = df.groupby('region')['cost'].mean().sort_values(ascending=False)

print("Average cost by brand : ")
print(avg_cost_by_brand)
print("\n")
print("Average cost by region : ")
print(avg_cost_by_region)
```

```python

# Heatmap by brand and region
pivot = df.pivot_table(index='brand', columns='region', values='cost', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Average Cost by Brand and Region")
plt.tight_layout()
plt.show()
```
